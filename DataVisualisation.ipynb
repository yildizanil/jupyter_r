## Global temperature anomalies

**Data:**

GISTEMP Team, 2021: GISS Surface Temperature Analysis (GISTEMP), version 4. NASA Goddard Institute for Space Studies. Dataset accessed 2021-05-22 at [data.giss.nasa.gov/gistemp/](https://data.giss.nasa.gov/gistemp/")

Lenssen, N., G. Schmidt, J. Hansen, M. Menne, A. Persin, R. Ruedy, and D. Zyss, 2019: Improvements in the GISTEMP uncertainty model. J. Geophys. Res. Atmos., 124, no. 12, 6307-6326, doi: [10.1029/2018JD029522](https://doi.org/10.1029/2018JD029522).

```r
global_temp <- read.table(
    "https://data.giss.nasa.gov/gistemp/graphs/graph_data/Global_Mean_Estimates_based_on_Land_and_Ocean_Data/graph.txt",
    skip = 5)
col_global_temp <- rep(NA, nrow(global_temp))
col_global_temp[which(global_temp[, 2] < 0)] <- "#c5e9e7"
col_global_temp[which(global_temp[, 2] > 0)] <- "#ffa891"
head(global_temp)
```

```r
options(repr.plot.width = 8, repr.plot.height = 5)
par(mar = c(3.5, 3.5, 0.25, 0.25), mgp = c(0.1, 0.1, 0),
    ps = 14, family = "serif", las = 1, pty = "m")
plot(0, 0, xlim = c(1880, 2021), ylim = c(-1.1, 1.1),
    xlab = NA, ylab = NA, axes = F)
segments(x0 = seq(1880, 2020, 10), y0 = -1.1,
    x1 = seq(1880, 2020, 10), y1 = 1.1, col = "gray87")
segments(x0 = 1880, y0 = seq(-1.1, 1.1, 0.1), x1 = 2020,
    y1 = seq(-1.1, 1.1, 0.1), col = "gray87")
segments(x0 = 1880, y0 = 0, x1 = 2020, y1 = 0, col = 1)
axis(1, tck = 0.02)
axis(2, tck = 0.02)
box()
lines(global_temp[, 2]~global_temp[, 1], lwd = 2, col = 1)
par(las = 0)
mtext("Year", side = 1, line = 1.5)
mtext("Temperature anomaly [Â°C]", side = 2, line = 2)
```

```r
options(repr.plot.width = 8, repr.plot.height = 5)
par(mar = c(3.5, 3.5, 0.25, 0.25), mgp = c(0.1, 0.1, 0),
    ps = 14, family = "serif", las = 1, pty = "m")
plot(0, 0, xlim = c(1880, 2021), ylim = c(-1.1, 1.1),
    xlab = NA, ylab = NA, axes = F)
segments(x0 = seq(1880, 2020, 10), y0 = -1.1,
    x1 = seq(1880, 2020, 10), y1 = 1.1, col = "gray87")
segments(x0 = 1880, y0 = seq(-1.1, 1.1, 0.1),
    x1 = 2020, y1 = seq(-1.1, 1.1, 0.1), col = "gray87")
segments(x0 = 1880, y0 = 0,x1 = 2020, y1 = 0, col = 1)
axis(1, tck = 0.02)
axis(2, tck = 0.02)
box()
points(global_temp[, 2]~global_temp[, 1], lwd = 1, pch = 16)
par(las = 0)
mtext("Year", side = 1, line = 1.5)
mtext("Temperature anomaly [Â°C]", side = 2, line = 2)
```

```r
positives <- global_temp[global_temp[, 2] > 0, ]
negatives <- global_temp[global_temp[, 2] < 0, ]
```

```r
options(repr.plot.width = 8, repr.plot.height = 5)
par(mar = c(3.5, 4, 0.25, 0.25), mgp = c(0.1, 0.1, 0),
    ps = 14, family = "serif", las = 1, pty = "m")
plot(0, 0, xlim = c(1880, 2021), ylim = c(-1.1, 1.1),
    xlab = NA, ylab = NA, axes = F)
segments(x0 = seq(1880, 2020, 10), y0 = -1.1,
    x1 = seq(1880, 2020, 10), y1 = 1.1, col = "gray87")
segments(x0 = 1880, y0 = seq(-1.1, 1.1, 0.1),
    x1 = 2020, y1 = seq(-1.1, 1.1, 0.1), col = "gray87")
segments(x0 = 1880, y0 = 0, x1 = 2020, y1 = 0, col = 1)
axis(1, tck = 0.02)
axis(2, tck = 0.02)
box()
lines(global_temp[, 2]~global_temp[, 1], lwd = 2, col = 1)
points(negatives[, 2]~negatives[, 1], lwd = 1, bg = "#c5e9e7", pch = 21)
points(positives[, 2]~positives[, 1], lwd = 1, bg = "#ffa891", pch = 21)
par(las = 0)
mtext("Year", side = 1, line = 1.5)
mtext("Temperature anomaly [Â°C]", side = 2, line = 2)
```

```r
options(repr.plot.width = 8, repr.plot.height = 5)
par(mar = c(3.5, 4, 0.25, 0.25), mgp = c(0.1, 0.1, 0),
    ps = 14, family = "serif", las = 1, pty = "m")
plot(0, 0, xlim = c(1880, 2020), ylim = c(-1.1, 1.1),
    xlab = NA, ylab = NA, axes = F)
axis(1, tck = 0.02)
axis(2, tck = 0.02)
box()
polygon(x = c(global_temp[, 1], rev(global_temp[, 1])),
    y = c(ifelse(global_temp[, 2] < 0, global_temp[, 2], 0), rep(0, 141)),
    border = NA, col = "#c5e9e7")
polygon(x = c(global_temp[, 1], rev(global_temp[, 1])),
    y = c(ifelse(global_temp[, 2] > 0, global_temp[, 2], 0), rep(0, 141)),
    border = NA, col = "#ffa891")
par(las = 0)
mtext("Year", side = 1, line = 1.5)
mtext("Temperature anomaly [Â°C]", side = 2, line = 2)
```

### Warming stripes by Ed Hawkins

**Method for visualisation:** Warming stripes by Ed Hawkins <a href="https://www.climate-lab-book.ac.uk/2018/warming-stripes/">climate-lab-book.ac.uk/2018/warming-stripes/</a>\
**Data:** GISTEMP Team, 2021: GISS Surface Temperature Analysis (GISTEMP), version 4. NASA Goddard Institute for Space Studies. Dataset accessed 2021-05-22 at <a href="https://data.giss.nasa.gov/gistemp/">data.giss.nasa.gov/gistemp/</a>.\
Lenssen, N., G. Schmidt, J. Hansen, M. Menne, A. Persin, R. Ruedy, and D. Zyss, 2019: Improvements in the GISTEMP uncertainty model. J. Geophys. Res. Atmos., 124, no. 12, 6307-6326, doi: <a href="https://doi.org/10.1029/2018JD029522">10.1029/2018JD029522</a>.




```r
library(RColorBrewer)
breaks <- seq(-0.5, 1.1, length.out = 12)
options(repr.plot.width = 10, repr.plot.height = 5)
par(mar = c(0, 0, 0, 0), mgp = c(0.1, 0.1, 0),
   ps = 14, family = "serif", las = 1, pty = "m")
plot(0, 0, xlim = c(1880, 2021), ylim = c(0, 1),
   xlab = NA, ylab = NA, axes = F)
for (i in 1:141) {
   rect(xleft = 1880 + (i - 1), ybottom = 0,
      xright = 1880 + i, ytop = 1,
      col = brewer.pal(n = 11, name = "RdBu")[12 - findInterval(global_temp[i, 2], breaks)],
      border = NA)
}
```

## Air temperature in Newcastle upon Tyne, UK

**Data:** Yildiz, A. & Stirling, R. 2020. PLEXUS Lysimeter Meteorological Data. doi: <a href="https://doi.org/10.25405/data.ncl.12789323.v3">10.25405/data.ncl.12789323.v3</a>.

```r
airtemp <- read.csv("https://data.ncl.ac.uk/ndownloader/files/24194087")
airtemp[, 1] <- as.POSIXct(airtemp[, 1], tz = "UTC", format = "%d/%m/%Y %H:%M")
summary(airtemp)
```

```r
start <- as.POSIXct("2019-05-03 00:00", tz = "UTC")
end <- as.POSIXct("2019-07-18 00:00", tz = "UTC")
```

```r
time_steps_24h <- seq(start, end, 60 * 60 * 24)
which_step_24h <- findInterval(as.POSIXct(airtemp[, 1], tz = "UTC"),
    time_steps_24h)
airtemp_24h <- aggregate(airtemp[, 2], list(time_steps_24h[which_step_24h]),
    mean, na.rm = T)
airtemp_max <- aggregate(airtemp[, 2], list(time_steps_24h[which_step_24h]),
    max, na.rm = T)
airtemp_min <- aggregate(airtemp[, 2], list(time_steps_24h[which_step_24h]),
    min, na.rm = T)
axis_steps_24h <- seq(start + 60 * 60 * 12, end, 60 * 60 * 24)
head(airtemp_24h)
```

```r
axis_days <- seq(start, end, 60 * 60 * 24)
axis_weeks <- seq(start, end, 60 * 60 * 24 * 7)
axis_week_names <- paste0(substr(axis_weeks, 9, 10), "-",
    substr(axis_weeks, 6, 7))
```

```r
options(repr.plot.width = 10, repr.plot.height = 5)
par(mar = c(3.5, 4, 0.25, 0.25), mgp = c(0.1, 0.1, 0),
    ps = 14, family = "serif", las = 1, pty = "m")
plot(0, 0, xlim = c(start, end), ylim = c(0, 30),
    xlab = NA, ylab = NA, axes = F)
segments(x0 = axis_days, y0 = 0,
    x1 = axis_days, y1 = 30,
    col = "gray87")
segments(x0 = start, y0 = seq(0, 30, 5),
    x1 = end, y1 = seq(0, 30, 5), col = "gray87")
axis(1, tck = 0.01, at = axis_days, labels = NA)
axis(1, tck = 0.02, at = axis_weeks, labels = axis_week_names)
axis(2, tck = 0.02)
box()
lines(airtemp$AirTemp~airtemp$Time, lwd = 2)
par(las = 0)
mtext("Time [dd-mm-2019]", side = 1, line = 2)
mtext("Air temperature [Â°C]", side = 2, line = 2)
```

```r
options(repr.plot.width = 10, repr.plot.height = 5)
par(mar = c(3.5, 4, 0.25, 0.25), mgp = c(0.1, 0.1, 0),
    ps = 14, family = "serif", las = 1, pty = "m")
plot(0, 0, xlim = c(start, end), ylim = c(0, 30),
    xlab = NA, ylab = NA, axes = F)
segments(x0 = axis_days, y0 = 0,
    x1 = axis_days, y1 = 30,
    col = "gray87")
segments(x0 = start, y0 = seq(0, 30, 5),
    x1 = end, y1 = seq(0, 30, 5),
    col = "gray87")
axis(1, tck = 0.01, at = axis_days, labels = NA)
axis(1, tck = 0.02, at = axis_weeks, labels = axis_week_names)
axis(2, tck = 0.02)
box()
lines(airtemp$AirTemp~airtemp$Time, lwd = 2, col = 8)
points(airtemp_24h[, 2]~axis_steps_24h, col = 2, pch = 16)
par(las = 0)
mtext("Time [dd-mm-2019]", side = 1, line = 2)
mtext("Air temperature [Â°C]", side = 2, line = 2)
```

```r
options(repr.plot.width = 10, repr.plot.height = 5)
par(mar = c(3.5, 4, 0.25, 0.25), mgp = c(0.1, 0.1, 0),
    ps = 14, family = "serif", las = 1, pty = "m")
plot(0, 0, xlim = c(start, end), ylim = c(0, 30),
    xlab = NA, ylab = NA, axes = F)
segments(x0 = axis_days, y0 = 0,
    x1 = axis_days, y1 = 30,
    col = "gray87")
segments(x0 = start, y0 = seq(0, 30, 5),
    x1 = end, y1 = seq(0, 30, 5),
    col = "gray87")
axis(1, tck = 0.01, at = axis_days, labels = NA)
axis(1, tck = 0.02, at = axis_weeks, labels = axis_week_names)
axis(2, tck = 0.02)
box()
lines(airtemp_24h[, 2]~axis_steps_24h, col = 2, lwd = 2)
points(airtemp_24h[, 2]~axis_steps_24h, col = 2, pch = 16)
par(las = 0)
mtext("Time [dd-mm-2019]", side = 1, line = 2)
mtext("Air temperature [Â°C]", side = 2, line = 2)
```

```r
options(repr.plot.width = 10, repr.plot.height = 5)
par(mar = c(3.5, 4, 0.25, 0.25), mgp = c(0.1, 0.1, 0),
    ps = 14, family = "serif", las = 1, pty = "m")
plot(0, 0, xlim = c(start, end), ylim = c(0, 30),
    xlab = NA, ylab = NA, axes = F)
segments(x0 = axis_days, y0 = 0,
    x1 = axis_days, y1 = 30,
    col = "gray87")
segments(x0 = start, y0 = seq(0, 30, 5),
    x1 = end, y1 = seq(0, 30, 5),
    col = "gray87")
axis(1, tck = 0.01, at = axis_days, labels = NA)
axis(1, tck = 0.02, at = axis_weeks, labels = axis_week_names)
axis(2, tck = 0.02)
box()
lines(airtemp_24h[, 2]~axis_steps_24h, col = 2, lwd = 2)
points(airtemp_24h[, 2]~axis_steps_24h, bg = 2, pch = 21)
polygon(x = c(axis_steps_24h, rev(axis_steps_24h)),
    y = c(airtemp_max[, 2], rev(airtemp_min[, 2])),
    col = rgb(1, 0, 0, 0.3), border = NA)
par(las = 0)
mtext("Time [dd-mm-2019]", side = 1, line = 2)
mtext("Air temperature [Â°C]", side = 2, line = 2)
```

## Energy consumption in Leipzig

**Data:** <a href="https://statistik.leipzig.de/opendata/api/values?kategorie_nr=13&rubrik_nr=4&periode=y&format=csv">Energieverbrauch (Jahreszahlen)</a> by <a href="https://opendata.leipzig.de/">Stadt Leipzig</a> is licensed under <a href="https://www.govdata.de/dl-de/by-2-0">DL-DE/BY-2.0.</a>
 <br>

```r
leipzig <- read.csv(
    "https://statistik.leipzig.de/opendata/api/values?kategorie_nr=13&rubrik_nr=4&periode=y&format=csv")
head(leipzig)
leipzig <- leipzig[seq(3, 12, 2), 1:17]
```

```r
data <- as.matrix(leipzig[, 3:17])
for (i in 1:15) {
 data[, i] <- 100 * data[, i] / sum(data[, i])
 }
 head(data)
```

```r
options(repr.plot.width = 10, repr.plot.height = 5)
layout(matrix(c(1, 2), nrow = 2, ncol = 1), widths = c(16), heights = c(1, 7))
par(mar = c(0, 0, 0, 0), mgp = c(0.1, 0.1, 0),
       ps = 14, family = "serif", las = 1, pty = "m")
plot(0, 0, type = "l", ylim = c(0, 50), xlim = c(2000, 2014),
       pch = "", axes = F, xlab = NA, ylab = NA)
legend("center",
       c("Electricity", "District heating", "Natural gas",
              "Coal", "Heating oil"),
       ncol = 3, lwd = 2,
       col = c("#ffcc91", "#ef7b61", "#c64022", "#afc6f3", "#31738a"),
       bty = "n", pch = c(1, 0, 16, 8, 15))
par(mar = c(3.5, 4, 0.25, 0.25), mgp = c(0.1, 0.1, 0),
       ps = 14, family = "serif", las = 1, pty = "m")
plot(0, 0, type = "l", ylim = c(0, 50), xlim = c(2000, 2014),
       pch = "", axes = F, xlab = NA, ylab = NA)
segments(x0 = seq(2000, 2014, 1), y0 = 0,
       x1 = seq(2000, 2014, 1), y1 = 50, col = "gray87")
segments(x0 = 2000, y0 = seq(0, 50, 5),
       x1 = 2014, y1 = seq(0, 50, 5), col = "gray87")
axis(1, tck = 0.02)
axis(2, tck = 0.02)
lines(data[1, ]~seq(2000, 2014, 1), lwd = 2,
       col = "#ffcc91", type = "b", pch = 1)
lines(data[2, ]~seq(2000, 2014, 1), lwd = 2,
       col = "#ef7b61", type = "b", pch = 0)
lines(data[3, ]~seq(2000, 2014, 1), lwd = 2,
       col = "#c64022", type = "b", pch = 16)
lines(data[4, ]~seq(2000, 2014, 1), lwd = 2,
       col = "#afc6f3", type = "b", pch = 8)
lines(data[5, ]~seq(2000, 2014, 1), lwd = 2,
       col = "#31738a", type = "b", pch = 15)
par(las = 0)
mtext("Year", side = 1, line = 1.5)
mtext("Share of energy consumption [%]", side = 2, line = 2)
box()
```
